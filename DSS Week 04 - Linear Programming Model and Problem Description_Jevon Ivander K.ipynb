```python
# Install library or-tools jika belum ada
!pip install ortools

from ortools.linear_solver import pywraplp

def SolveProductionPlan():
    # 1. Inisialisasi Solver menggunakan GLOP (Google Linear Optimization Package)
    solver = pywraplp.Solver.CreateSolver('GLOP')
    if not solver:
        return

    # 2. Variabel Keputusan
    # x1: Produk 1, x2: Produk 2, x3: Produk 3
    x1 = solver.NumVar(0, solver.infinity(), 'Produk_1')
    x2 = solver.NumVar(0, solver.infinity(), 'Produk_2')
    x3 = solver.NumVar(0, solver.infinity(), 'Produk_3')

    print('Jumlah variabel =', solver.NumVariables())

    # 3. Fungsi Kendala (Constraints) berdasarkan kapasitas mesin per hari
    # Kendala Mesin M1: 2x1 + 3x2 + 2x3 <= 440
    solver.Add(2 * x1 + 3 * x2 + 2 * x3 <= 440)

    # Kendala Mesin M2: 4x1 + 3x3 <= 470
    solver.Add(4 * x1 + 3 * x3 <= 470)

    # Kendala Mesin M3: 2x1 + 5x2 <= 430
    solver.Add(2 * x1 + 5 * x2 <= 430)

    # 4. Fungsi Tujuan (Objective Function): Memaksimalkan Profit
    # Profit: Rs 4 (x1), Rs 3 (x2), Rs 6 (x3)
    solver.Maximize(4 * x1 + 3 * x2 + 6 * x3)

    # 5. Memanggil Solver
    status = solver.Solve()

    # 6. Menampilkan Hasil
    if status == pywraplp.Solver.OPTIMAL:
        print('Solusi Ditemukan:')
        print(f'Keuntungan Maksimum = Rs. {solver.Objective().Value():.2f}')
        print(f'Produksi Produk 1 = {x1.solution_value():.2f} unit')
        print(f'Produksi Produk 2 = {x2.solution_value():.2f} unit')
        print(f'Produksi Produk 3 = {x3.solution_value():.2f} unit')
    else:
        print('Masalah tidak dapat diselesaikan secara optimal.')

SolveProductionPlan()
```

```python
# Install library or-tools jika belum ada
# !pip install ortools

from ortools.linear_solver import pywraplp

def SolvePerusahaanA():
    # 1. Inisialisasi Solver
    solver = pywraplp.Solver.CreateSolver('GLOP')
    if not solver:
        return

    # 2. Perhitungan Kapasitas (Konversi Satuan)
    # Bahan Baku (kg ke ons: 1 kg = 10 ons)
    kapasitas_A = 120 * 10  # 1200 ons
    kapasitas_B = 90 * 10   # 900 ons
    kapasitas_C = 125 * 10  # 1250 ons

    # Waktu Kerja (Jam per Minggu)
    # 08.00-16.00 (8 jam) - 1 jam istirahat = 7 jam efektif/hari
    jam_efektif_hari = 7
    hari_kerja = 6
    kapasitas_Proses_I = 10 * jam_efektif_hari * hari_kerja  # 420 jam
    kapasitas_Proses_II = 12 * jam_efektif_hari * hari_kerja # 504 jam

    # 3. Variabel Keputusan
    x1 = solver.NumVar(0, solver.infinity(), 'Produk_I')
    x2 = solver.NumVar(0, solver.infinity(), 'Produk_II')
    x3 = solver.NumVar(0, solver.infinity(), 'Produk_III')
    x4 = solver.NumVar(0, solver.infinity(), 'Produk_IV')

    # 4. Fungsi Kendala (Constraints)
    # Kendala Bahan Baku
    solver.Add(10*x1 + 8*x2 + 6*x3 + 9*x4 <= kapasitas_A) # Bahan A
    solver.Add(6*x1 + 10*x2 + 8*x3 + 5*x4 <= kapasitas_B)  # Bahan B
    solver.Add(12*x1 + 9*x2 + 5*x3 + 6*x4 <= kapasitas_C) # Bahan C

    # Kendala Proses Produksi
    solver.Add(4*x1 + 3*x2 + 2*x3 + 6*x4 <= kapasitas_Proses_I)  # Proses I
    solver.Add(2*x1 + 4*x2 + 3*x3 + 5*x4 <= kapasitas_Proses_II) # Proses II

    # 5. Fungsi Tujuan (Maximize Profit)
    solver.Maximize(2000*x1 + 1900*x2 + 1600*x3 + 2100*x4)

    # 6. Menjalankan Solver
    status = solver.Solve()

    # 7. Output Hasil
    if status == pywraplp.Solver.OPTIMAL:
        print('--- Hasil Optimasi Produksi Perusahaan A ---')
        print(f'Keuntungan Maksimum = Rp{solver.Objective().Value():,.0f}')
        print(f'Produksi Produk I   = {x1.solution_value():.2f} unit')
        print(f'Produksi Produk II  = {x2.solution_value():.2f} unit')
        print(f'Produksi Produk III = {x3.solution_value():.2f} unit')
        print(f'Produksi Produk IV  = {x4.solution_value():.2f} unit')
    else:
        print('Solusi optimal tidak ditemukan.')

SolvePerusahaanA()
```

```python
# Install library ortools jika belum ada
# !pip install ortools

from ortools.linear_solver import pywraplp

def SelesaikanOptimasiMebel():
    # 1. Inisialisasi Solver menggunakan GLOP (Google Linear Optimization Package)
    solver = pywraplp.Solver.CreateSolver('GLOP')
    if not solver:
        return

    # 2. Variabel Keputusan
    # x: Jumlah Meja, y: Jumlah Kursi
    # Kita asumsikan jumlah produksi tidak bisa negatif (>= 0)
    x = solver.NumVar(0, solver.infinity(), 'Meja')
    y = solver.NumVar(0, solver.infinity(), 'Kursi')

    print('Jumlah variabel =', solver.NumVariables())

    # 3. Fungsi Kendala (Constraints) berdasarkan jam kerja tersedia
    # Kendala Perakitan: 4x + 2y <= 60
    solver.Add(4 * x + 2 * y <= 60)

    # Kendala Pemolesan: 2x + 4y <= 48
    solver.Add(2 * x + 4 * y <= 48)

    # 4. Fungsi Tujuan (Objective Function): Memaksimalkan Laba
    # Laba Meja = 80.000, Laba Kursi = 60.000
    solver.Maximize(80000 * x + 60000 * y)

    # 5. Memanggil Solver
    status = solver.Solve()

    # 6. Menampilkan Hasil
    if status == pywraplp.Solver.OPTIMAL:
        print('\n--- Solusi Optimal Ditemukan ---')
        print(f'Jumlah Meja yang harus dibuat = {x.solution_value():.0f} unit')
        print(f'Jumlah Kursi yang harus dibuat = {y.solution_value():.0f} unit')
        print(f'Laba Maksimal Total = Rp {solver.Objective().Value():,.0f}')
    else:
        print('Masalah tidak dapat diselesaikan secara optimal.')

SelesaikanOptimasiMebel()
```

```python
# Install library ortools jika belum ada
# !pip install ortools

from ortools.linear_solver import pywraplp

def SelesaikanMinimasiRoyal():
    # 1. Inisialisasi Solver menggunakan GLOP (Linear Programming)
    solver = pywraplp.Solver.CreateSolver('GLOP')
    if not solver:
        return

    # 2. Variabel Keputusan
    # x: Jumlah Royal Bee, y: Jumlah Royal Jelly
    x = solver.NumVar(0, solver.infinity(), 'Royal_Bee')
    y = solver.NumVar(0, solver.infinity(), 'Royal_Jelly')

    print('Jumlah variabel =', solver.NumVariables())

    # 3. Fungsi Kendala (Constraints)
    # Kendala Vitamin: 2x + 1y >= 8
    solver.Add(2 * x + 1 * y >= 8)

    # Kendala Protein: 2x + 3y >= 12
    solver.Add(2 * x + 3 * y >= 12)

    # Kendala Produksi Minimal
    solver.Add(x >= 2) # Royal Bee minimal 2
    solver.Add(y >= 1) # Royal Jelly minimal 1

    # 4. Fungsi Tujuan (Objective Function): Meminimumkan Biaya
    # Biaya Royal Bee = 100, Biaya Royal Jelly = 80
    solver.Minimize(100 * x + 80 * y)

    # 5. Memanggil Solver
    status = solver.Solve()

    # 6. Menampilkan Hasil
    if status == pywraplp.Solver.OPTIMAL:
        print('\n--- Solusi Minimasi Biaya Ditemukan ---')
        print(f'Jumlah Royal Bee   = {x.solution_value():.2f} unit')
        print(f'Jumlah Royal Jelly = {y.solution_value():.2f} unit')
        print(f'Total Biaya Minimum = Rp {solver.Objective().Value():,.2f}')
    else:
        print('Masalah tidak dapat diselesaikan secara optimal.')

SelesaikanMinimasiRoyal()
```

```python
# Install library ortools jika belum ada
# !pip install ortools

from ortools.linear_solver import pywraplp

def SelesaikanCampuranLogam():
    # 1. Inisialisasi Solver
    solver = pywraplp.Solver.CreateSolver('GLOP')
    if not solver:
        return

    # 2. Variabel Keputusan (Proporsi A, B, dan C dalam 1 ton campuran)
    xA = solver.NumVar(0, 1, 'Bahan_A')
    xB = solver.NumVar(0, 1, 'Bahan_B')
    xC = solver.NumVar(0, 1, 'Bahan_C')

    print('Jumlah variabel =', solver.NumVariables())

    # 3. Fungsi Kendala (Constraints)

    # Kendala Total Proporsi (xA + xB + xC = 1)
    solver.Add(xA + xB + xC == 1)

    # Kendala Specific Gravity (<= 0.98)
    solver.Add(0.92 * xA + 0.97 * xB + 1.04 * xC <= 0.98)

    # Kendala Chromium (>= 8%)
    solver.Add(7 * xA + 13 * xB + 16 * xC >= 8)

    # Kendala Melting Point (>= 450 C)
    solver.Add(440 * xA + 490 * xB + 480 * xC >= 450)

    # 4. Fungsi Tujuan (Objective Function): Meminimumkan Biaya
    # Biaya: A=90, B=280, C=40
    solver.Minimize(90 * xA + 280 * xB + 40 * xC)

    # 5. Memanggil Solver
    status = solver.Solve()

    # 6. Menampilkan Hasil
    if status == pywraplp.Solver.OPTIMAL:
        print('\n--- Solusi Optimal Ditemukan ---')
        print(f'Proporsi Bahan A = {xA.solution_value():.2f} (50%)')
        print(f'Proporsi Bahan B = {xB.solution_value():.2f} (0%)')
        print(f'Proporsi Bahan C = {xC.solution_value():.2f} (50%)')
        print(f'Total Biaya Minimum per Ton = Rs {solver.Objective().Value():,.2f}')
    else:
        print('Masalah tidak dapat diselesaikan secara optimal.')

SelesaikanCampuranLogam()
```
